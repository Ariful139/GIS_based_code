```python
import os
import rasterio
from rasterio.enums import Resampling
import numpy as np
```

```python
# Base directory containing Mosaic folders
base_directory = "D:/BUET/Thesis/ArcGIS/Mosaiced Landsat"  # Replace with your actual path
output_directory = "D:/BUET/Thesis/ArcGIS/NDWI2"  # Replace with your desired output path

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)
```

```python
# Loop through each year from 1988 to 2024
for year in range(1988, 2025):
    folder_name = f"Mosaic_{year}"
    folder_path = os.path.join(base_directory, folder_name)

    # Check if the folder exists
    if not os.path.isdir(folder_path):
        print(f"{folder_name} folder does not exist.")
        continue

    # Define band paths based on year and NDWI equation
    if year <= 2013:
        band2_path = os.path.join(folder_path, f"Mosaic_{year}_B2.tif")
        band4_path = os.path.join(folder_path, f"Mosaic_{year}_B4.tif")

        # Ensure both files exist
        if not (os.path.isfile(band2_path) and os.path.isfile(band4_path)):
            print(f"Missing bands for year {year}. Skipping.")
            continue

        # Open Band 2 and capture profile info
        with rasterio.open(band2_path) as band2_src:
            band2 = band2_src.read(1)
            profile = band2_src.profile

        # Open Band 4 and align dimensions if needed
        with rasterio.open(band4_path) as band4_src:
            with rasterio.vrt.WarpedVRT(band4_src, width=band2.shape[1], height=band2.shape[0], 
                                        transform=profile['transform'], resampling=Resampling.nearest) as band4_vrt:
                band4 = band4_vrt.read(1)

        # Calculate NDWI
        ndwi = (band2 - band4) / (band2 + band4)

    else:
        band3_path = os.path.join(folder_path, f"Mosaic_{year}_B3.tif")
        band5_path = os.path.join(folder_path, f"Mosaic_{year}_B5.tif")

        # Ensure both files exist
        if not (os.path.isfile(band3_path) and os.path.isfile(band5_path)):
            print(f"Missing bands for year {year}. Skipping.")
            continue

        # Open Band 3 and capture profile info
        with rasterio.open(band3_path) as band3_src:
            band3 = band3_src.read(1)
            profile = band3_src.profile

        # Open Band 5 and align dimensions if needed
        with rasterio.open(band5_path) as band5_src:
            with rasterio.vrt.WarpedVRT(band5_src, width=band3.shape[1], height=band3.shape[0], 
                                        transform=profile['transform'], resampling=Resampling.nearest) as band5_vrt:
                band5 = band5_vrt.read(1)

        # Calculate NDWI
        ndwi = (band3 - band5) / (band3 + band5)

    # Save NDWI result
    output_file_path = os.path.join(output_directory, f"NDWI_{year}.tif")
    profile.update(dtype=rasterio.float32, count=1)

    with rasterio.open(output_file_path, 'w', **profile) as dst:
        dst.write(ndwi.astype(rasterio.float32), 1)

    print(f"NDWI for {year} saved to {output_file_path}")
```
